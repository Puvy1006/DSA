```python
# ------------------------------------------------------------
# Cleaning NBA Player Stats (Handling Outliers and Ranking)
# ------------------------------------------------------------
# Steps:
#   1. Read nba_player_stats.csv
#   2. Show summary statistics
#   3. Detect and fix outliers in performance columns
#   4. Display top 10 players by PER (Player Efficiency Rating)
# ------------------------------------------------------------

import pandas as pd
import numpy as np

# Step 1: Read dataset
filename = "nba_player_stats.csv"   # Ensure file is in the same folder
data = pd.read_csv(filename)

# Step 2: Show summary statistics
print("----- Summary Statistics -----\n")
print(data.describe())

# Step 3: Handle outliers in numeric performance columns
performance_cols = ["Points", "Assists", "Rebounds", "Steals", "Blocks"]  # adjust names to match your file

for col in performance_cols:
    if col in data.columns:
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1

        lower_limit = Q1 - 1.5 * IQR
        upper_limit = Q3 + 1.5 * IQR

        # Cap outliers to upper/lower limits
        data[col] = np.where(data[col] > upper_limit, upper_limit,
                      np.where(data[col] < lower_limit, lower_limit, data[col]))
        print(f"\nâœ… Fixed outliers in '{col}' (Capped between {lower_limit:.2f} and {upper_limit:.2f})")

# Step 4: Show how many missing or extreme values are now fixed
print("\n----- Cleaned Data Summary -----\n")
print(data[performance_cols].describe())

# Step 5: Show top 10 players by PER
if "PER" in data.columns:
    print("\nğŸ† Top 10 Players by Player Efficiency Rating (PER):\n")
    top_10 = data.sort_values(by="PER", ascending=False).head(10)
    print(top_10[["Player", "Team", "PER", "Points", "Assists", "Rebounds"]])
else:
    print("\nâš  'PER' column not found in dataset!")

# Step 6: Save cleaned dataset
data.to_csv("nba_player_stats_cleaned.csv", index=False)
print("\nâœ… Cleaned dataset saved as 'nba_player_stats_cleaned.csv'")
```
