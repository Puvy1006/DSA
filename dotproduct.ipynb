```python
import numpy as np

# Take number of rows and columns
rows = int(input("Enter number of rows: "))
cols = int(input("Enter number of columns: "))

# Input for Matrix A
print("\nEnter elements for Matrix A:")
A = []
for i in range(rows):
    row = list(map(int, input(f"Row {i+1}: ").split()))
    A.append(row)

# Input for Matrix B
print("\nEnter elements for Matrix B:")
B = []
for i in range(rows):
    row = list(map(int, input(f"Row {i+1}: ").split()))
    B.append(row)

# Convert lists to NumPy arrays
A = np.array(A)
B = np.array(B)

# Display matrices
print("\nMatrix A:\n", A)
print("\nMatrix B:\n", B)

# Element-wise operations
print("\n--- Element-wise Operations ---")
print("A + B:\n", A + B)
print("\nA - B:\n", A - B)
print("\nA * B (Element-wise Multiplication):\n", A * B)
print("\nA / B (Element-wise Division):\n", A / B)

# Matrix operations
print("\n--- Matrix Operations ---")
if A.shape[1] == B.shape[0]:
    print("Dot Product (A @ B):\n", np.dot(A, B))
else:
    print("Dot Product not possible (Columns of A â‰  Rows of B)")

print("\nTranspose of A:\n", A.T)
print("Shape of A:", A.shape)
print("Shape of A after Transpose:", A.T.shape)

# Using arange(), shape, and reshape()
print("\n--- Using arange(), shape, and reshape() ---")
arr = np.arange(1, 13)
print("Array created with arange(1,13):\n", arr)
print("Shape of arr:", arr.shape)

reshaped = arr.reshape(3, 4)
print("\nReshaped array (3x4):\n", reshaped)
print("Shape after reshape:", reshaped.shape)
```
