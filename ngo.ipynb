```python
import operator

# Sample volunteer data
volunteers = [
    {"vid": 1, "name": "Alice", "program": "Education", "hours": {"Q1": 10, "Q2": 12, "Q3": 15, "Q4": 8}, "city": "CityA", "year": 2024},
    {"vid": 2, "name": "Bob", "program": "Health", "hours": {"Q1": 5, "Q2": 7, "Q3": 6, "Q4": 10}, "city": "CityB", "year": 2024},
    {"vid": 3, "name": "Charlie", "program": "Education", "hours": {"Q1": 12, "Q2": 15, "Q3": 10, "Q4": 10}, "city": "CityA", "year": 2025},
    {"vid": 4, "name": "Diana", "program": "Environment", "hours": {"Q1": 8, "Q2": 9, "Q3": 12, "Q4": 10}, "city": "CityC", "year": 2024},
    {"vid": 5, "name": "Eve", "program": "Health", "hours": {"Q1": 10, "Q2": 8, "Q3": 12, "Q4": 9}, "city": "CityB", "year": 2025},
    {"vid": 6, "name": "Frank", "program": "Education", "hours": {"Q1": 7, "Q2": 10, "Q3": 12, "Q4": 10}, "city": "CityA", "year": 2024},
    # Add more volunteer entries as needed
]

# ----------------------------
# 1. Average hours for each program
# ----------------------------
program_totals = {}
program_counts = {}

for v in volunteers:
    program = v["program"]
    total_hours = sum(v["hours"].values())
    if program not in program_totals:
        program_totals[program] = 0
        program_counts[program] = 0
    program_totals[program] += total_hours
    program_counts[program] += 1

print("Average Hours per Program:")
for program in program_totals:
    avg = program_totals[program] / program_counts[program]
    print(f"{program}: {avg:.2f} hours")

# ----------------------------
# 2. Top 10 volunteers with most yearly hours
# ----------------------------
for v in volunteers:
    v["total_hours"] = sum(v["hours"].values())

# Sort by total_hours descending
top_volunteers = sorted(volunteers, key=lambda x: x["total_hours"], reverse=True)[:10]

print("\nTop 10 Volunteers by Yearly Hours:")
for v in top_volunteers:
    print(f"{v['name']} ({v['program']}) - {v['total_hours']} hours")

# ----------------------------
# 3. Volunteer retention rate
# ----------------------------
# Count how many volunteers from a previous year continued the next year
volunteers_by_year = {}
for v in volunteers:
    year = v["year"]
    if year not in volunteers_by_year:
        volunteers_by_year[year] = set()
    volunteers_by_year[year].add(v["vid"])

years = sorted(volunteers_by_year.keys())
retained = 0
if len(years) > 1:
    previous_year_volunteers = volunteers_by_year[years[0]]
    next_year_volunteers = volunteers_by_year[years[1]]
    retained = len(previous_year_volunteers & next_year_volunteers)
    retention_rate = (retained / len(previous_year_volunteers)) * 100
    print(f"\nVolunteer Retention Rate from {years[0]} to {years[1]}: {retention_rate:.2f}%")
else:
    print("\nNot enough data to calculate retention rate.")
```
