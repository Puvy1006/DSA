```python
import pandas as pd

# Step 1: Read dataset
filename = "weather_aus.csv"   # Make sure this file is in the same folder
data = pd.read_csv(filename)

# Step 2: Show summary statistics
print("----- Summary Statistics -----\n")
print(data.describe())

# Step 3: Detect outliers in 'Rainfall' column using IQR method
Q1 = data['Rainfall'].quantile(0.25)
Q3 = data['Rainfall'].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper limits
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

print("\nIQR for Rainfall:", IQR)
print("Lower Limit:", lower_limit)
print("Upper Limit:", upper_limit)

# Step 4: Find outliers
outliers = data[(data['Rainfall'] < lower_limit) | (data['Rainfall'] > upper_limit)]
print("\nNumber of outliers detected in Rainfall:", len(outliers))

# Step 5: Remove outliers
data_removed = data[(data['Rainfall'] >= lower_limit) & (data['Rainfall'] <= upper_limit)]
print("\nRows after removing outliers:", len(data_removed))
print("Rows removed:", len(data) - len(data_removed))

# Step 6: Save cleaned dataset
data_removed.to_csv("weather_cleaned_removed.csv", index=False)

print("\n Cleaned data saved as 'weather_cleaned_removed.csv' (outliers removed)")
```
