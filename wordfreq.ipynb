```python
# Word Frequency Analysis in Social Media Text
from collections import Counter
import matplotlib.pyplot as plt
import string

# Step 1: Read the text file
with open("social_media.txt", "r", encoding="utf-8") as file:
    text = file.read().lower()  # Convert text to lowercase

# Step 2: Remove punctuation
translator = str.maketrans('', '', string.punctuation)
cleaned_text = text.translate(translator)

# Step 3: Split text into words
words = cleaned_text.split()

# Step 4: Count word frequencies
word_counts = Counter(words)

# Step 5: Get the 10 most common words
top_10 = word_counts.most_common(10)
print("Top 10 words with counts:", top_10)

# Step 6: Separate words and counts using list comprehension
words_list = [item[0] for item in top_10]
counts_list = [item[1] for item in top_10]

# Step 7: Plot the bar graph
plt.figure(figsize=(10,6))
plt.bar(words_list, counts_list, color='skyblue')
plt.title("Top 10 Most Used Words in social_media.txt")
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.show()

# Step 8: Easy Summary
print("Summary: The most frequent word is '{}' with {} occurrences.".format(top_10[0][0], top_10[0][1]))
```
