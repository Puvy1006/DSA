```python
import numpy as np

# Take number of rows and columns
rows = int(input("Enter the number of rows: "))
cols = int(input("Enter the number of columns: "))

# Input for Matrix A
print("\nEnter the elements for Matrix A:")
A = []
for i in range(rows):
    row = list(map(int, input(f"Row {i+1}: ").split()))
    A.append(row)

# Input for Matrix B
print("\nEnter the elements for Matrix B:")
B = []
for i in range(rows):
    row = list(map(int, input(f"Row {i+1}: ").split()))
    B.append(row)

# Convert lists to NumPy arrays
A = np.array(A)
B = np.array(B)

# Display matrices
print("\nMatrix A:\n", A)
print("\nMatrix B:\n", B)

# Operations
print("\nMatrix Addition:\n", A + B)
print("\nMatrix Subtraction:\n", A - B)
print("\nMatrix Multiplication (element-wise):\n", A * B)
print("\nMatrix Division (element-wise):\n", A / B)
print("\nMatrix Power (A ** 2):\n", A ** 2)

# Other NumPy functions
print("\nMatrix of Zeros (3x3):\n", np.zeros((3, 3)))
print("\nMatrix of Ones (2x4):\n", np.ones((2, 4)))

# Shape, Reshape, and arange()
print("\nShape of A:", A.shape)
arr = np.arange(1, 13)
print("\nArray using arange(1, 13):\n", arr)
print("\nReshaped array (3x4):\n", arr.reshape(3, 4))
```
